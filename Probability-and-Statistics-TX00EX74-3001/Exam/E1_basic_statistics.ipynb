## Setup

Import the necessary packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Example 1

We create a numeric array and calculate the [basic statistics](https://numpy.org/doc/stable/reference/routines.statistics.html):
- mean
- median
- standard deviation
- min and max
- percentiles

```python
# Example data
x = np.array([7, 0, 7, 4, 8, 8, 2, 9, 0, 9])
print(x)
```

```python
# Mean
m = np.mean(x)
print(f'mean = {m:.2f}')
```

```python
# Sorted values and median
me = np.median(x)
print(f'sorted values = ', np.sort(x))
print(f'median = {me:.2f}')
```

```python
# Standard deviation
sd = np.std(x)
print(f'std = {sd:.2f}')
```

```python
# Min and max
xmin, xmax = np.min(x), np.max(x)
print(f'min and max = {xmin:.2f} and {xmax:.2f}')
```

```python
# Percentiles
p = np.percentile(x, [25, 50, 75])
print('Sorted values = ', np.sort(x))
print(f'25% = {p[0]:.2f}')
print(f'50% = {p[1]:.2f}')
print(f'75% = {p[2]:.2f}')
```

## Example 2 - Basic statistics with pandas

Pandas library offers tools to import and handle *tabular data*. Here are examples how to import csv (comma separated values) data and calculate the basic statistics.


```python
dataframe = pd.read_csv('example2.csv')
dataframe
```

```python
# Show the first 5 rows
dataframe.head()
```

```python
# Show the last five rows
dataframe.tail()
```

```python
# Calculate the mean values
dataframe.mean()
```

```python
# Calculate the standard deviations
dataframe.std()
```

```python
# Calculate the mean value of the first column
m = dataframe["col1"].mean()
print(m)
```

```python
# Count how many value are in first column
c = dataframe["col1"].count()
print(c)
```

```python
# Calculate the 25%, 50%, and 75% percentiles for the second column
col2 = dataframe["col2"]
p = np.percentile(col2, [25, 50, 75])
print(p)
```

```python
# Calculate the descriptive statistics to whole dataframe
dataframe.describe()
```

```python
# Show all possible unique values for column 1
v = dataframe["col1"].unique()
print(v)
```

```python
# Sort the values
v_sorted = np.sort(v)
print(v_sorted)
```

## Example 3 - Basic graphical presentations of data

In the following example we demonstrate how to visualize the data. This example shows how to:

- Select a column from a pandas dataframe
- Use line-plot, scatter-plot, or stem-plot to visualize all data points
- Plot sorted values
- Use histogram, boxplot, or violin-plot to graph the distribution of the values


```python
# Select the first column from dataframe
x1 = dataframe['col1']
```

### Plot, scatter, and stem

```python
# Plot the values
plt.plot(x1)
plt.xlabel('Index')
plt.ylabel('x1 values')
plt.title('Line plot of the x1 values')
plt.show()
```

```python
# Scatter plot of the x1 values
index = np.arange(len(x1))
plt.scatter(index, x1)
plt.xlabel('Index')
plt.ylabel('x1 values')
plt.title('Scatter plot of x1 values')
plt.show()
```

```python
# Stem-plot of the values
plt.stem(x1, orientation = 'horizontal', bottom = 4.34)
plt.title('Stem-plot of x1 values')
plt.xlabel('Value of x1')
plt.ylabel('Index')
plt.show()
```

### Graphing sorted values

It is visually easier to read the sorted data values. Here are examples of showing the stem-plot and line-plot of sorted values of x1.  

```python
# Stem-plot of the sorted values
plt.stem(np.sort(x1), orientation = 'horizontal', bottom = 4.34)
plt.title('Stem-plot of sorted x1 values')
plt.xlabel('Value of x1')
plt.ylabel('Index (sorted)')
plt.show()
```

```python
# Line-plot of the sorted values
plt.plot(np.sort(x1))
plt.xlabel('Index')
plt.ylabel('x1 values')
plt.title('Line plot of the sorted x1 values')
plt.show()
```

### Hist, boxplot, and violinplot

Histograms, boxplots, and violinplots are used to illustrate the distributions of the values. Here are examples how to call these functions.

```python
# Histrogram of the values
plt.hist(x1, bins = 10)
plt.xlabel('Value of x1')
plt.ylabel('Count')
plt.title('Histogram of x1 values')
plt.show()
```

```python
# Boxplot of x1 values
plt.boxplot(x1)
plt.title('Boxplot of x1 values')
plt.ylabel('Value of x1')
plt.grid()
plt.show()
```

```python
# Violin-plot of x1 values
plt.violinplot(x1)
plt.title('Violin-plot of x1 values')
plt.ylabel('Value of x1')
plt.show()
```

### Visualizing relationship

Lastly, let's see how to visualize relationship between two variables. Most common technique is to use scatter-plot, where we plot the first value on x-direction and the second value on y-direction. See this example.

```python
x1 = dataframe['col1']
x2 = dataframe['col2']
plt.scatter(x1, x2)
plt.title('Scatter plot of x2 vs. x1')
plt.xlabel('Value of x1')
plt.ylabel('Value of x2')
plt.grid()
plt.show()
```
