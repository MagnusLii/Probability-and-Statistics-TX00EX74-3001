```python
# (a) Histograms
import matplotlib.pyplot as plt

df = pd.read_csv('CSVs/emg3087.csv')

# Plot histogram for force column
plt.hist(df['force'], bins=10)
plt.xlabel('Force')
plt.ylabel('Frequency')
plt.title('Histogram of clench force')
plt.show()

# Plot histogram for emg column
plt.hist(df['emg'], bins=10)
plt.xlabel('EMG')
plt.ylabel('Frequency')
plt.title('Histogram of EMG')
plt.show()
```

```python
# calculate the linear regression line
x = df['force']
y = df['emg']
m, b = np.polyfit(x, y, 1)

# scattergram
plt.scatter(x, y)
plt.xlabel('Clench Force')
plt.ylabel('EMG')
plt.title('Scattergram of Clench Force and EMG')

# plot the linear regression line
plt.plot(x, m*x + b, color='red')

plt.show()
```

```python
# Correlation between the signals
correlation = df.corr()
print(correlation)

```

**Interpretation:**<br>

* (e) While not a perfect correlation the two values seem to correlate enough to use one value to predict the other with a fairly hight degree of confidence. <br>

```python
df = pd.read_csv('CSVs/heartrate688.csv')

plt.hist(df['garmin'], bins=10, alpha=0.5, label='Garmin')
plt.hist(df['cosinuss'], bins=10, alpha=0.5, label='Cosinuss')
plt.hist(df['polar'], bins=10, alpha=0.5, label='Polar')
plt.hist(df['firstbeat'], bins=10, alpha=0.5, label='Firstbeat')

plt.xlabel('Heart Rate')

fig, axs = plt.subplots(2, 2, figsize=(10, 16))

# Garmin histogram
axs[0, 0].hist(df['garmin'], bins=10, alpha=0.5)
axs[0, 0].set_xlabel('Heart Rate')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].set_title('Garmin')

# Cosinuss histogram
axs[0, 1].hist(df['cosinuss'], bins=10, alpha=0.5)
axs[0, 1].set_xlabel('Heart Rate')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].set_title('Cosinuss')

# Polar histogram
axs[1, 0].hist(df['polar'], bins=10, alpha=0.5)
axs[1, 0].set_xlabel('Heart Rate')
axs[1, 0].set_ylabel('Frequency')
axs[1, 0].set_title('Polar')

# Firstbeat histogram
axs[1, 1].hist(df['firstbeat'], bins=10, alpha=0.5)
axs[1, 1].set_xlabel('Heart Rate')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].set_title('Firstbeat')

plt.show()
```

```python
df = pd.read_csv('CSVs/heartrate688.csv')

plt.scatter(df['time'],df['garmin'],  label='Garmin')
plt.scatter(df['time'], df['cosinuss'],  label='Cosinuss')
plt.scatter( df['time'], df['polar'], label='Polar')
plt.scatter( df['time'], df['firstbeat'], label='Firstbeat')
plt.xlabel('Device')
plt.ylabel('Time')
plt.title('Scatterplots of Device Readings vs Time')
plt.legend()

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Garmin scatterplot
axs[0, 0].scatter(df['time'], df['garmin'])
axs[0, 0].set_xlabel('Time')
axs[0, 0].set_ylabel('Garmin')
axs[0, 0].set_title('Scatterplot of Garmin vs Time')

# Cosinuss scatterplot
axs[0, 1].scatter(df['time'], df['cosinuss'])
axs[0, 1].set_xlabel('Time')
axs[0, 1].set_ylabel('Cosinuss')
axs[0, 1].set_title('Scatterplot of Cosinuss vs Time')

# Polar scatterplot
axs[1, 0].scatter(df['time'], df['polar'])
axs[1, 0].set_xlabel('Time')
axs[1, 0].set_ylabel('Polar')
axs[1, 0].set_title('Scatterplot of Polar vs Time')

# Firstbeat scatterplot
axs[1, 1].scatter(df['time'], df['firstbeat'])
axs[1, 1].set_xlabel('Time')
axs[1, 1].set_ylabel('Firstbeat')
axs[1, 1].set_title('Scatterplot of Firstbeat vs Time')

plt.tight_layout()
plt.show()

```

```python
# Looping through device readings and counting how many consecutive values are the same.
# A high number might suggest a slow polling/reading rate.

devices = ['garmin', 'cosinuss', 'polar', 'firstbeat']

for device in devices:
    consecutive_count = 0
    previous_value = None
    
    for value in df[device]:
        if value == previous_value:
            consecutive_count += 1
        previous_value = value
    
    print(f"Consecutive count for {device}: {consecutive_count}")
```
